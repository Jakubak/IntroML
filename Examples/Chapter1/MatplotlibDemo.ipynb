# Matplotlib Demo
For more details about **Matplotlib**, click [here](https://matplotlib.org/3.2.2/contents.html) to check out the documentation.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Basics

```python
# list of integers from 0 to 19
x1 = [i for i in range(20)]
x1
```

```python
y1 = x1
y1
```

```python
# plot of y=x, the 'r-' argument results in a solid red line, set canvas size to 5x3 pixels
plt.figure(figsize=(5,3))
plt.plot(x1, y1, 'r-')
plt.title('Plot of y=x')
plt.xlabel('x label')
plt.ylabel('y label')

plt.show()
```

```python
# 20 equally spaced points between -5 and 5
x2 = np.linspace(-5, 5, 20)
x2
```

```python
# y = x^2
y2 = x2 ** 2
```

```python
# plot of y=x^2, the 'bo' argument results in a blue dotted line
plt.figure()
plt.plot(x2, y2, 'bo')
plt.title('Plot of y=x^2')
plt.xlabel('x label')
plt.ylabel('y label')

plt.show()
```

## Dealing with row vectors

```python
x3 = x2.reshape(1, len(x2))
x3
```

```python
y3 = np.exp(x3)
y3
```

```python
plt.figure()

plt.plot(np.squeeze(x3), np.squeeze(y3), 'g-')
plt.title('Plot of y=e^x')

plt.show()
```

### Object Oriented Approach

```python
# initialize figure object
fig = plt.figure()

# add set of axes to figure 
ax = fig.add_axes([0, 0, 1, 1])

# plot x^2 and x^3 on same canvas with legend
ax.plot(x2, x2**2, 'bo', label='X^2')
ax.plot(x2, x2**3, 'r-', label='X^3')
ax.legend()

plt.show()
```

## Subplots

```python
plt.figure()
# 1 row 2 columns first plot
plt.subplot(1, 2, 1)
plt.plot(x1, y1, color='orange', linestyle='-', linewidth=6, alpha=0.5)

# 1 row 2 columns second plot
plt.subplot(1, 2, 2)
plt.plot(x2, y2, color='green', marker='o', linewidth=0.5)

plt.show()
```

### Object Oriented Approach

```python
# canvas with 1 row and 2 columns
fig, axes = plt.subplots(nrows=1, ncols=2)

# plot y=x on the left
axes[0].plot(x1, y1, color='orange', linestyle='-', linewidth=6, alpha=0.5)
axes[0].set_title('Plot of y=x')
axes[0].set_xlabel('x label')
axes[0].set_ylabel('y label')

# plot y=x^2 on the right
axes[1].plot(x2, y2, color='green', marker='o', linewidth=0.5)
axes[1].set_title('Plot of y=x^2')
axes[1].set_xlabel('x label')

fig
```

## Scatter plot

```python
# set seed for reproducibility
np.random.seed(10)

# create a 2x50 matrix with random normally distributed data
X = np.random.randn(2, 50)

X
```

```python
plt.figure()

# plot the data in X where the points in the first array are the x coordinates 
# and in the second array are the corresponding y coordinates
x_coords = X[0]
y_coords = X[1]
plt.scatter(x_coords, y_coords)

plt.show()
```

```python
plt.figure()

# extract the points to the left of x=0
left_x = x_coords[x_coords < 0]
left_y = y_coords[x_coords < 0]

# extract the points to the right of x=0
right_x = x_coords[x_coords >= 0]
right_y = y_coords[x_coords >= 0]

plt.scatter(left_x, left_y, color='blue', label='class 0')
plt.scatter(right_x, right_y, color='red', label='class 1')
plt.legend()

plt.show()
```

## Scatter plot + line

```python
# set seed for reproducibility
np.random.seed(10)

# create a vector of 20 equally spaced points between 0 and 9
x = np.linspace(0, 10, 20)

# y is a linear function of x
y = 2*x + 1

# generate normally distributed random noise with mean 0 and standard dev 3
noise = np.random.normal(0, 3, 20)

# add some noise to data points
y_data = y + noise

# plot data points
plt.scatter(x, y_data)

# plot true function (line)
plt.plot(x, y, 'r-')

plt.show()
```

## Histogram

```python
# set seed for reproducibility
np.random.seed(10)

# create a vector with 100 random normally distributed numbers
normal_vector = np.random.randn(100)

# plot normally distributed vector using a histogram
plt.hist(normal_vector)

plt.show()
```

## Barchart

```python
# set seed for reproducibility
np.random.seed(10)

# create a vector of 5 random floats in the interval [0, 1)
probabilities = np.random.random_sample(5)

days = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday']

plt.figure()

# plot vertical bar chart
plt.bar(x=days, height=probabilities, color='orange')
plt.title('Probability of Studying On a Certain Day of the Week')
plt.xlabel('Day')
plt.ylabel('Probability')

plt.show()
```

```python
# alternative way
fig, ax = plt.subplots()
ax.bar(x=days, height=probabilities, color='#6B8E23', alpha=0.6)
ax.set_title('Probability of Studying On a Certain Day of the Week')
ax.set_ylabel('Probability')
ax.set_xlabel('Day')

plt.show()
```

## Heatmap

### Simple Example

```python
# create a 100x100 matrix with all 0s to the left and all 1s to the right
heatmap_mask = np.zeros((100, 100))
heatmap_mask[:, 50:] = 1

heatmap_mask
```

```python
plt.figure()

# plot matrix as a heatmap with legend bar
plt.pcolormesh(heatmap_mask, cmap='RdBu')
plt.colorbar()

plt.show()
```

### Intermediate Example

```python
npoints = 100

# create 100 equally spaced points between -3 and 3
X0lin = np.linspace(-3, 3, npoints)

# create 100 equally spaced points between -5 and 5
X1lin = np.linspace(-5, 5, npoints)

X0lin
```

```python
# create grids for heatmap to define axes range
X0grid, X1grid = np.meshgrid(X0lin, X1lin)

print("X0grid.shape: {}".format(X0grid.shape))
X0grid
```

```python
plt.figure()

# plot the grids with the given mask
plt.pcolormesh(X0grid, X1grid, heatmap_mask, cmap='RdBu')
plt.xlabel('X0')
plt.ylabel('X1')
plt.colorbar()

plt.show()
```

```python
plt.figure()

# set seed for reproducibility
np.random.seed(10)

# create a 2x100 matrix with random normally distributed data
X = np.random.randn(2, 100)

# let x0 coordinates be the first array of X
x0_coords = X[0]

# let x1 coordinates be the second array of X
x1_coords = X[1]

# extract the points to the left of X0=0
left_x0 = x0_coords[x0_coords < 0]
left_x1 = x1_coords[x0_coords < 0]

# extract the points to the right of X0=0
right_x0 = x0_coords[x0_coords >= 0]
right_x1 = x1_coords[x0_coords >= 0]


# plot the grids with the given mask
plt.pcolormesh(X0grid, X1grid, heatmap_mask, cmap='RdBu')
plt.xlabel('X0')
plt.ylabel('X1')
plt.colorbar()

# plot the points
plt.scatter(left_x0, left_x1, color='red', label='class 0')
plt.scatter(right_x0, right_x1, color='blue', label='class 1')
plt.legend()

plt.show()
```

### Advanced Example

```python
def func_Y(X):
    # generate function of 0 and 1: Y = 0 if X0 + X1-0.75<=0 and Y=1 if x0+x1-0.75>0
    Y = Y = (X[0,:] + X[1,:] - 0.75 > 0).astype(float)
    Y = np.expand_dims(Y,axis=0)
    print("X.shape: {}".format(X.shape))
    print("Y.shape: {}".format(Y.shape))
    return Y
```

```python
# define training data dimensions
nfeature = 2
m = 1000

# create a random normally distributed 2x1000 array of synthetic training data
X = np.random.randn(nfeature, m)

# map the values in X to a label
Y = func_Y(X)

Y
```

```python
plt.figure()

# get indices where Y is 0 
idx_0 = np.squeeze(Y) < 1

# get indices where Y is 1
idx_1 = np.squeeze(Y) >= 1

# plot red points
plt.scatter(X[0,idx_0], X[1,idx_0], color='red', label="Training Data: Label=0")

# plot blue points
plt.scatter(X[0,idx_1], X[1,idx_1], color='blue', label="Training Data: Label=1")

# place legend at bottom right corner
plt.legend(loc='lower right')

plt.xlabel('X0')
plt.ylabel('X1')

plt.show()
```

```python
# let x0 coordinates be the first array of X
x0_coords = X[0]

# let x1 coordinates be the second array of X
x1_coords = X[1]

npoints = 100

# create 1d grids in x0 and x1 directions, add 1 to max and subtract 1 from min
# so that points are inside heatmap
X0lin = np.linspace(np.min(x0_coords)-1, np.max(x0_coords)+1, npoints)
X1lin = np.linspace(np.min(x1_coords)-1, np.max(x1_coords)+1, npoints)

# create 2d grads for x0 and x1 and reshape into 1d grids 
X0grid,X1grid = np.meshgrid(X0lin, X1lin)
X0reshape = np.reshape(X0grid, (1,npoints*npoints))
X1reshape = np.reshape(X1grid, (1,npoints*npoints))

# Xmatrix is a 2d grid (2,npoints*npoints)
Xmatrix = np.concatenate((X0reshape, X1reshape), axis=0)
Yout = func_Y(Xmatrix)

# reshape results into 2d grid and plot heatmap
heatmap_mask = np.reshape(Yout, (npoints, npoints))
plt.pcolormesh(X0grid, X1grid, heatmap_mask, cmap='RdBu')
plt.colorbar()

plt.show()
```

```python
plt.figure()

# plot the heatmap
plt.pcolormesh(X0grid, X1grid, heatmap_mask, cmap='RdBu')
plt.colorbar()

# plot red and blue points
plt.scatter(X[0,idx_0], X[1,idx_0], color='red', label="Training Data: Label=0")
plt.scatter(X[0,idx_1], X[1,idx_1], color='blue', label="Training Data: Label=1")
plt.xlabel('X0')
plt.ylabel('X1')
plt.legend()

plt.show()
```
