```python
import sklearn as sk
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

## Reading and Inspecting the Data

```python
data = pd.read_csv('../Data/Train_Fares_train.csv').dropna()

data.head()
```

```python
data.describe()
```

```python
data.info()
```

```python
nrows = data.shape[0]
ncols = data.shape[1]

print(f'The dataset consists of {nrows} rows and {ncols} columns.')
```

```python
# calculate total fare
fare_components = ['tolls_amount', 'tip_amount', 'mta_tax', 'surcharge', 'fare_amount']
data['total_fare'] = data[fare_components].sum(axis=1)
```

## Exploratory Data Analysis

```python
def plot_distribution(df, column):
    """Plot the distribution of a given column in a dataframe."""
    plt.figure()
    plt.hist(df[column].values)
    plt.title(f'Distribution of {column}')
    plt.xlabel(column)
    plt.ylabel('Count')
    
    plt.show()
```

```python
# get the numerical features
numerical_features = [col for col in data.columns if data[col].dtypes != 'O']
numerical_features
```

```python
for feature in numerical_features:
    plot_distribution(data, feature)
```
